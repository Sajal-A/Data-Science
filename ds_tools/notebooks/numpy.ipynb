<b>What is the major reason behind numpy's faster computation?</b>
<li>The numpy arrays is densely packed in memory due to it's homogeneous type.</li>
<li>NumPy is able to divide a task into multiple subtasks and process them parallely.</li>
<li>NumPy functions are implmented in C, which again makes it faster than python list</li>
<br>
<b>In Conclusion:</b>
<li>Numpy provides clean syntax for element-wise operation.</li>
<li>Per loop time for numpy to perform is much lesser than list</li>
<li>for large scale numerical computations, leveraing Numpy can result in significant performance improvments</li>

```python
import time
nums = list(range(0,1000000))

#case:1 - looping over a list
#Get the start time
st = time.time()
sq_nums = []
for i in nums:
    sq_nums.append(nums[i]**2) #Find the square of each number
#Get the end time
et = time.time()

#get the execution time
print('Execution time using loops over list: ',et-st)


#case:2 - using list comprehension
#Get the start time
st = time.time()
sq_nums = [x**2 for x in nums]
#Get the end time
et = time.time()

#get the execution time
print('Execution time using list comprehension: ',et-st)


#case3: Using numpy array broadcasting
numpy_num = np.arange(1000000)
#Get the start time
st = time.time()
numpy_num**2
#Get the end time
et = time.time()

#get the execution time
print('Execution time using numpy array broadcasting: ',et-st)
```

<b>UseCase: AirBnB NPS</b>

<b>Analyze user survey data and report NPS to the management</b>
<br>We gnerally encountered this question -> HOW LIKELY IS IT THAT YOU WOULD RECOMMEND [COMPANY X] TO A FRIEND OR COLLEAGUE? 
<br>In the range [0....10] --> 0(Not at all likely) -- 10(Extremely likely)
<br><b>This is called Likelyhood to Recommend Survey</b>
<br>Based on this we calculate the Net Promoter Score(NPS)


<b>How to calculate NPS score?</b>
<br>We label our response into 3 categories:
<li>Detractors: Respondents with a score of 0-6.</li>
<li>Passive: Respondents with a score of 7-8.</li>
<li>Promoters: Score of 9-10</li>

<b>Net Promoter Score = % Promoters - % Detractors</b>


<b>How is NPS helpful ?</b>
<br>NPS helps a brand in guaging its brand value and sentiment in the market.
<li>Promoters are highly likely to recommend your product or service. Hence, bringing in more business.</li>
<li>Whereas, Detractors are likely to recommend against your product or service's usage. Hence, bringing the business down.</li>
These insights can help business make customer oriented decision along with product improvisation.

<b>Interpreting NPS</b>
<li>NPS can range from -100 (if every customer is a detrator) to +100 (if every customer is a promoter).</li>
<li>A positve NPS(>0) is generally considered good, while an NPS of +50 is excellent</li>



<b>About the Dataset</b>
<p>
    These dataset focuses on user feedback received by Airbnb,containng
    1167 entries with the rating ranging from 0 to 10. These feedback
    entries provide valuable insights into user experiences and satisfaction levels.

</p>

```python
import numpy as np 
score = np.loadtxt('survey.txt',dtype='int') #load data from a text file, with aim to be a fast reader for simple text files.
```

```python
#let's see how the data looks like.
score[:10]
```

```python
#let's check the number of responses
score.shape 
```

There are a total of 1167 responses for the survey.

```python
#Let's check some sanity check on data
#Basic statistical analysis

#let's check the minimum rating for the survey
print('Minimum rating: ',score.min())

#let's check the maximum rating for the survey
print('Maximum rating: ',score.max())

#find the most occuring rating 
(unique, counts) = np.unique(score,return_counts=True) #returns the unique ratings and their respective counts.
most_common_rating = unique[np.argmax(counts)] #np.argmax()-- return the index of the maximum number in the array.

print(f'The most occurring rating is: {most_common_rating}')
print('Rating counts: ',dict(zip(unique,counts)))
```

<b>Let's Calculate the NPS</b>

```python
''' 
    % detractors
'''
#How can we get the count of detractors?
#using fancy indexing(masking)
detractors = score[score<=6].shape[0] #number shows the rating upto 6

#total count
total = score.shape[0]

#calculate the detractors pecentage
percent_detractors = (detractors/total)*100

print('Count of Detractors: ',detractors)
print('percent of detractors: ',percent_detractors)
```

```python
''' 
    % promoterd
'''
#How can we get the count of detractors?
#using fancy indexing(masking)
promoters = score[score>=9].shape[0] #number shows the rating >= 9

#total count
total = score.shape[0]

#calculate the detractors pecentage
percent_promoters= (promoters/total)*100

print('Count of Detractors: ',promoters)
print('percent of detractors: ',percent_promoters)
```

```python
''' 
    Calculating the NPS
'''
nps = round(percent_promoters - percent_detractors,ndigits=2)
print(nps)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
img = np.array(plt.imread('fruit.jpg'))
plt.imshow(img)
```

```python
print('Dimension of the image: ',img.ndim)
print('Shape of the image: ',img.shape)
```

<b>How does a 2d image has become 3 dimensional?</b><br>
1. Coloured images have a 3rd dimension for depth or RGB colour channel
2. Here, the depth is 3
<br>
<b>The shape of the image we read: (804, 1200, 3)</b><br>
1. matplotlib first reads that each plane has 804 * 1200 pixels.
2. then it reads depth value(R,G,B values) of each plane and place the values in 3 seperate planes.
3. That is why depth us the last element of shape tuple in np.array generated from an image and read by matplotlib.
4. Whereas in a normal np.array, depth is the first element of shape tuple.

<b>Visualizing RGB Channels</b>

We can split the image into each RGB color channels using Numpy.<br>
<b>What Exactly RGB values are?</b>
1. These are values of each pixel of an image.
2. Each pixel is made up of 3 components/channels - Red, Green, Blue - which form RGB values.
3. Coloured images are usually stored as 3-dimensional arrays of 8-but unsigned integers.
4. So, the range of values that each channel of a pixel can take is 0 to 2^8 - 1 (ie, can range from 0 to 255)
<br>Each pixel has these 3 values which combined together forms the color that the pixel represents<br>
1. So, a pixel[255,0,0] will be RED in color.
2. a pixel [0,255,0] will be GREEN in color.
3. a pixel [0,0,255] will be BLUE in color.
<br>Note: [0,0,0] represents black and [255,255,255] represents white


```python
#let's seperate the R, G, B channels in our image:
img_r = img.copy()
img_r[:,:,(1,2)]=0 #For RED channel, we'll set values of GREEN and BLUE to 0
#img_r[:,:,...] means that all rows and all columns of the image are being selected in each of the first two dimension
#(1,2) in the third dimension refers to the second and third channels of the image.
# the assignment sets the values of the selected channels to 0
#channel 0: Red, channel 1:Green, channel 2:Blue
```

```python
plt.imshow(img_r)
```

##### Array Splitting & Merging

<b>Splitting</b> - np.split()<br>
- Splits an array into multiple sub-arrays as views
- It takes an argument indices_or_sections(n) - (the array will be divided into n equal arrays along axis.)
- if such a split is not possible, an error is raised

```python
x = np.arange(12)
x
```

```python
np.split(x,4)
```

```python
#if indices_or_sections is a 1-d array of sorted integers, the entries indicate where along axis the array is split
#if an index exceeds the dimension of the array along axis, an empty sub-array is returned
np.split(x,[3,6,12])
```

<b>np.expand_dims()</b><br>
- Expands the shape of an array wit the axis of length 1.
- Insert a new axis that will appear at the position in the expanded array shape
- np.expand_dims(arr,axis)

```python
arr_1 = np.arange(0,6)
arr_1
```

```python
arr_1.shape
```

```python
#let's expand the dimension
expand_arr1 = np.expand_dims(arr_1,axis=0)
print(expand_arr1)
print(expand_arr1.shape)
```

- Here, the shape of array(6,) i.e only have one axis (axis=0)
- <b>When we expand dimension with axis = 0</b>
    - it add 1 to dimension @ axis = 0
    - shape becomes (1,6) from (6,)

```python
#let's expand dims @axis = 1
expand_arr2 = np.expand_dims(expand_arr1,axis=1)
expand_arr2, expand_arr2.shape 
```

##### Shallow Vs Deep Copy


```python
#let's take an example
arr_6 = np.arange(4)
arr_6 
```

```python
arr__6 = arr_6.reshape(2,2)
arr__6
```

```python
arr_6[0] = 232
arr_6
```

```python
arr__6
```

Array arr__6 got automatically updated when we change the value of arr_6.<br>This is called Shallow copy.
<b>What happens here?</b><br>
- Numpy re-uses data as much as possible instead of duplicating.
- This helps Numpy to be efficient.

<b>Now, let see an example where Numpy will create a 'Deep Copy' of data</b>


```python
a = np.arange(5)
a
```

```python
c = a+2
c
```

```python
a[0] = 23
a
```

```python
c #the content of the c remain same, i.e deep copy
```

- Numpy is able to use same data for simpler operation like reshape, slicing ---> <b>Shallow Copy</b>
- It creates a copy of data where operations make more permanent changes to data --> <b>Deep copy</b>

```python
#is there any way to check whether it share a same memory?
np.shares_memory(arr_6,arr__6)
```

<b>make shallow copy using .view() function</b><br>
- Any changes made in new array will be reflected in original array

```python
arr_7 = np.arange(10)
arr_7
```

```python
view_arr = arr_7.view()
view_arr
```

```python
view_arr[3] = 30
view_arr
```

```python
arr_7
```

<b>How to make deep copy? - using .copy() </b>

```python
arr_8 = arr_7.copy()
arr_8
```

```python
arr_7[5] =500
arr_7
```

```python
arr_8
```

<b>What is object arrays?</b><br>
- object arrays are basically array of any python datatype.
- .copy() will return shallow copy when copying the element of the array in object array.

```python
obj_arr = np.array([1,'m',[1,2,4]],dtype='object')
```

```python
obj_arr
```

<b>How to create deep copy then?</b>
- using copy.deepcopy()

```python
import copy

cop = copy.deepcopy(obj_arr)
cop
```

```python
cop[2][1] = 200
cop
```
